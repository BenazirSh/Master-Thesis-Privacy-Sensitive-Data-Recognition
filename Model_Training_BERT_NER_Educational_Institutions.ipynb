## Dataset loading and defining new labels


```python
# Import necessary libraries from HuggingFace for manipulating the dataset, and training the BERT transformer model
from datasets import load_dataset, ClassLabel, Sequence
from transformers import BertTokenizerFast, BertForTokenClassification, Trainer, TrainingArguments, DataCollatorForTokenClassification

# Load the CoNLL2003 dataset from HuggingFace.
# CoNLL2003 is a popular dataset I can use for Named Entity Recognition
# see: https://www.clips.uantwerpen.be/conll2003/ner/
dataset = load_dataset("conll2003")

print("Train dataset", len(dataset['train']))
print("Test dataset", len(dataset['test']))
print("Validation dataset", len(dataset['validation']))

# By default, the CoNLL2003 dataset has these labels: PER, ORG, LOC, MISC, O.
# Because I want to fine-tune the BERT model to
# recognize EDU labels for educational institutions (separate from ORG) in our CVs,
# I extend the label set here:
label_map = {
    0: 'O',
    1: 'B-PER',
    2: 'I-PER',
    3: 'B-ORG',
    4: 'I-ORG',
    5: 'B-LOC',
    6: 'I-LOC',
    7: 'B-MISC',
    8: 'I-MISC',
    9: 'B-EDU',  # Custom label for educational institutions
    10: 'I-EDU'  # Custom label for educational institutions
}

# Invert the label_map to create a mapping from label names to IDs for our convenience
label2id = {v: k for k, v in label_map.items()}
id2label = {k: v for k, v in label_map.items()}

# Update the dataset's features to account for the new labels
num_labels = len(label_map)  # Should be 11

# Modify the 'ner_tags' feature to be a sequence of ClassLabel
new_features = dataset["train"].features.copy()
new_features["ner_tags"] = Sequence(feature=ClassLabel(num_classes=num_labels, names=list(label_map.values())))

# Apply these new features to the dataset
dataset = dataset.cast(new_features)
new_features
```

## Relabeling the dataset

In this step I use [BIO tagging](https://en.wikipedia.org/wiki/Inside%E2%80%93outside%E2%80%93beginning_(tagging)) system to relabel the dataset.

I use the following algorithm:
1. Initialize a new list `new_labels` for every entry in the dataset.
1. Iterate over the `tokens` and `ner_tags` (labels) in a single loop, for every iteration where I have access to the tuple of a single `label` and `token`:
  * If the label for the current token is `B-ORG`, and if the token exists in the pre-defined list of 1500 universities, then I relabel it to `B-EDU`.
  * If the label for the current token is `I-ORG`, check the label at previous index `i-1`, and if it is `B-EDU` or `I-EDU` then I relabel the current label to `I-EDU`.
  * Otherwise, I keep the original label.

Lastly, I save the modified dataset to disk for convenience.

```python
# Define the relabeling function
def relabel_educational_institutions(example):
    new_labels = []
    for i, (label, token) in enumerate(zip(example['ner_tags'], example['tokens'])):
        if label == label2id['B-ORG'] and any([token.lower() in uni.lower() for uni in unis]):
            new_label = label2id['B-EDU']
        elif label == label2id['I-ORG'] and (new_labels[i-1] == label2id['B-EDU'] or new_labels[i-1] == label2id['I-EDU']):
            new_label = label2id['I-EDU']
        else:
            new_label = label  # Keep original label
        new_labels.append(new_label)
    example['ner_tags'] = new_labels
    return example

# Apply the relabeling function to the dataset
modified_dataset = dataset.map(relabel_educational_institutions)

modified_dataset.save_to_disk(f'./{RELABELED_DATASET_WITH_EDUCATIONAL_INSTITUTIONS}')
```

## Function to evaluate model metrics

Here I define a function `compute_metrics` that evaluates metrics (precision, recall, f1 and accuracy) for every epoch run during model training via Hugging Face (PyTorch).

In this case, as I am evaluating a Named-Entity-Recognition BERT model, for every input the model is defined to be accurate, iff every tag is accurate. If there is any incorrect predicted tag, then for that input the model's whole prediction is considered as inaccurate (wrong).

I filter out `-100` labels, because they are used to associate the special tokens `[CLS]` and `[SEP]` and they don't correspond to any NER classification target.

```python
import evaluate
import numpy as np

metric = evaluate.load('seqeval')

label_names = new_features['ner_tags'].feature.names

# This method will be called by Hugging Face after every epoch is completed during model training/fine-tuning process
# with a tuple parameter containing the predictions and labels.
def compute_metrics(eval_preds):
  logits, labels = eval_preds

  predictions = np.argmax(logits, axis=-1)

  # Filter out -100 as it is for special tokens [CLS] and [SEP]
  true_labels = [[label_names[l] for l in label if l != -100] for label in labels]

  true_predictions = [[label_names[p] for p,l in zip(prediction, label) if l != -100]
                      for prediction, label in zip(predictions, labels)]

  all_metrics = metric.compute(predictions=true_predictions, references=true_labels)

  return {"precision": all_metrics['overall_precision'],
          "recall": all_metrics['overall_recall'],
          "f1": all_metrics['overall_f1'],
          "accuracy": all_metrics['overall_accuracy']}
```

## Model training and fine-tuning using the relabeled dataset

This is the main step where I will be fine-tuning the model using the custom dataset (based on `CoNLL2003`) I prepared earlier.
1. I use the `bert-base-uncased` as the base for training the fine-tuned model.
1. Create the `tokenize_and_align_labels` method which is responsible for bringing labels and tokens to the *same length*, esecially it will be handling cases where there is *subword tokenization* (like "un", "##happi", "##ness").
1. I apply `DataCollatorForTokenClassification` to the tokenizer to make sure our tokens are padded to the length of the maximum length sequence in the batch.
1. I create an instance of `Trainer`, to which I also pass the `compute_metric` method, that is called by the trainer after every epoch, the tokenized_dataset and data collator.
1. At the end, I call the `trainer.train()` that kicks off the model training process and then call `save_model()` to save the final model to disk.

```python
BERT_BASE_MODEL = "bert-base-uncased"

# Initialize the tokenizer
tokenizer = BertTokenizerFast.from_pretrained(BERT_BASE_MODEL)

# Define the tokenization and label alignment function
def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer(examples['tokens'], truncation=True, is_split_into_words=True)

    # Initialize labels list to match the batched input
    labels = []

    for i, label_list in enumerate(examples['ner_tags']):  # Iterate over each example's labels (which is a list of lists)
        word_ids = tokenized_inputs.word_ids(batch_index=i)  # Get word IDs for the current batch index
        previous_word_idx = None
        label_ids = []

        # Iterate over each word in the example
        for word_idx in word_ids:
            if word_idx is None:
                # No label needed for special tokens, so I just add -100 for PyTorch
                label_ids.append(-100)
            elif word_idx != previous_word_idx:
                # Start of a new word, append the corresponding label
                label_ids.append(label_list[word_idx])  # Get the correct label for the current word
            else:
                # Inside a word, label should only be for the first token, hence ignore subsequent ones
                label_ids.append(label_list[word_idx] if id2label[label_list[word_idx]].startswith('I-') else -100)

            previous_word_idx = word_idx

        # Append the labels for the current example to the labels list
        labels.append(label_ids)

    tokenized_inputs["labels"] = labels
    return tokenized_inputs

# Apply the tokenization and label alignment
tokenized_dataset = modified_dataset.map(tokenize_and_align_labels, batched=True, remove_columns=dataset["train"].column_names)

# check tokenization
print("Tokenized sample:")
print(tokenized_dataset['train'][0])

# Initialize the model
model = BertForTokenClassification.from_pretrained(
    BERT_BASE_MODEL,
    num_labels=num_labels,  # Ensure num_labels is correctly set to 11
    id2label=id2label,
    label2id=label2id
)

# Define training arguments, note that here we set the number of epochs to 3
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    save_total_limit=2,
    logging_dir="./logs",
    logging_steps=10,
)

# Create a data collator
data_collator = DataCollatorForTokenClassification(tokenizer)

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["validation"],
    compute_metrics=compute_metrics,
    tokenizer=tokenizer,
    data_collator=data_collator,
)

# Start training
trainer.train()

# Save the model
trainer.save_model(f"./{FINE_TUNED_MODEL_WITH_EDUCATIONAL_INSTITUTIONS}")
```
